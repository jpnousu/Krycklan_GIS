```python
import os
import shutil

resos = [20, 40, 80, 160]

# Source and destination base directories
src_base = '/Users/jpnousu/Krycklan_GIS_data'
dst_base = '/Users/jpnousu/SpaFHy_RUNS/krycklan/gis'

# Mapping of source filenames -> destination filenames
file_map = {
    '5haStreams_depth.asc': '5haStreams_depth.asc',
    '5haStreams.asc': '5haStreams.asc',
    '5haStreams_length.asc': '5hStreams_length.asc',
    '5haStreams_distance.asc': '5haStreams_distance.asc',
    '5haStreams_width.asc': '5haStreams_width.asc',
    '5haStreams_area.asc': '5haStreams_area.asc',     
    
    'channels_depth.asc': 'channels_depth.asc',
    'channels.asc': 'channels.asc',
    'channels_length.asc': 'channels_length.asc',
    'channels_distance.asc': 'channels_distance.asc',
    'channels_width.asc': 'channels_width.asc',
    'channels_area.asc': 'channels_area.asc',       
    
    'canopy_fraction.asc': 'canopy_fraction.asc',
    'canopy_height.asc': 'canopy_height.asc',
    'cmask.asc': 'catchment_mask.asc',
    'cmask_bi.asc': 'catchment_mask_bi.asc',
    'dem_clip_no_deps.asc': 'processed_dem.asc',
    'krycklan_QD_J1_J2_final.asc': 'soil.asc',
    'LAI_decid.asc': 'LAI_decid.asc',
    'LAI_pine.asc': 'LAI_pine.asc',
    'LAI_spruce.asc': 'LAI_spruce.asc',
    'LAI_conif.asc': 'LAI_conif.asc',
    'lake_mask.asc': 'lake_mask.asc',
    'SOIL_DEPTH.asc': 'soildepth.asc',
    'twi_dinf.asc': 'twi_dinf.asc',
    'twi_d8.asc': 'twi_d8.asc',
    'flowacc_sca_dinf.asc': 'flow_accumulation_dinf.asc',
    'slope.asc': 'slope.asc'
}

for reso in resos:
    src_dir = os.path.join(src_base, f'{reso}m', 'ALL')
    dst_dir = os.path.join(dst_base, f'{reso}m')
    os.makedirs(dst_dir, exist_ok=True)
    
    for src_name, dst_name in file_map.items():
        src_path = os.path.join(src_dir, src_name)
        dst_path = os.path.join(dst_dir, dst_name)
        
        if os.path.exists(src_path):
            shutil.copy(src_path, dst_path)
            print(f"Copied {src_name} → {dst_name} for {reso}m")
        else:
            print(f"⚠️ File not found: {src_path}")
```
